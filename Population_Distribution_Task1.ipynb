```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_61.csv", skiprows=4)

df.head()
```

```python
data_2022 = df[['Country Name', '2022']]
data_2022 = data_2022.dropna()

data_2022.head()
```

```python
top10 = data_2022.sort_values(by='2022', ascending=False).head(10)

top10
```

```python
plt.figure(figsize=(10,6))

plt.bar(top10['Country Name'], top10['2022'])

plt.title("Top 10 Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.show()
```
